```python
def ConcatDf(date, cat):
    # date takes a string in the format 'YYYY-MM'
    # cat is 'stop-and-search' or 'street'
    files = glob('datasets\\' + date + '\\*-' + cat +'.csv') # get a list of the files in directory
    return pd.concat((pd.read_csv(file).assign(filename=file.replace('datasets\\' + date + '\\' + date + '-', ''))
                      for file in files), ignore_index=True)
```

```python
from datetime import datetime
dateList = [d.strftime('%Y-%m') for d in pd.date_range(start='2018-01', end='2020-12', freq='M')]
```

```python
dateList
```

```python
dfStreetList = [ConcatDf(date, 'street') for date in [d.strftime('%Y-%m') for d in pd.date_range(start='2018-01', end='2020-12', freq='M')]]
dfStreet = pd.concat(dfStreetList)
```

```python
dfStopList = [ConcatDf(date, 'stop-and-search') for date in [d.strftime('%Y-%m') for d in pd.date_range(start='2018-01', end='2020-12', freq='M')]]
dfStop = pd.concat(dfStopList)
```

```python
# Let's drop some irrelevant columns to reduce the size of our dataset. I'm doing this to make it easier to load whenever we start our app.
dfStreetCompact = dfStreet.drop(columns=['Crime ID', 'Reported by', 'Falls within', 'LSOA code', 'LSOA name', 'Context', 'filename'])
dfStopCompact = dfStop.drop(columns=['Part of a policing operation', 'Legislation', 'Outcome', 'Outcome linked to object of search', 'filename'])
```

```python
dfStreetCompact.to_csv('street_data.csv', index=False)
```
